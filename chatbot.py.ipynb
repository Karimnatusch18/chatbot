```python
import requests
import json
import time

# --- CONFIGURACIÓN ---
HF_TOKEN = "TU_TOKEN_AQUI" # <--- REEMPLAZA CON TU TOKEN
MODELO = "meta-llama/Meta-Llama-3-8B-Instruct"

# Probamos con la URL de router (si da 410, cambia a api-inference)
API_URL = f"https://router.huggingface.co/models/{MODELO}"
HEADERS = {"Authorization": f"Bearer {HF_TOKEN}"}

def llamar_modelo(mensaje, temperatura=0.7, max_tokens=500):
    payload = {
        "inputs": mensaje,
        "parameters": {
            "temperature": temperatura,
            "max_new_tokens": max_tokens,
            "return_full_text": False
        }
    }
    response = requests.post(API_URL, headers=HEADERS, json=payload)

    if response.status_code == 503:
        print("⏳ Monica se está preparando... espera 20 segundos.")
        time.sleep(20)
        response = requests.post(API_URL, headers=HEADERS, json=payload)

    if response.status_code == 200:
        return response.json()[0]["generated_text"].strip()
    else:
        return f"Error {response.status_code}: {response.text}"

print("✅ Celda 1: Motor de Monica preparado.")
```

```python
# ── MILESTONE 1: Define la personalidad ────
SYSTEM_PROMPT = """
Eres Monica, una entrenadora de inglés experta y muy motivadora.
Tu objetivo es ayudar al usuario a mejorar su inglés.
[RESTRICCIÓN 1]: Si el usuario escribe en inglés con errores, DEBES corregir la gramática primero y luego responder en español.
[RESTRICCIÓN 2]: Siempre sugiere una palabra de vocabulario más avanzada o un sinónimo para que el usuario aprenda.
[FORMATO]: Usa negritas para las correcciones y una sección llamada '💡 Tip de Vocabulario' al final de cada respuesta.
"""

print("✅ Celda 2: Personalidad de Monica definida.")
```

```python
# ── MILESTONE 2: Chatbot con personalidad ────
def chatbot_responder(pregunta_usuario, temperatura=0.7):
    prompt_completo = f"<|begin_of_text|><|start_header_id|>system<|end_header_id|>\n\n{SYSTEM_PROMPT}<|eot_id|><|start_header_id|>user<|end_header_id|>\n\n{pregunta_usuario}<|eot_id|><|start_header_id|>assistant<|end_header_id|>\n\n"
    return llamar_modelo(prompt_completo, temperatura)

# Prueba inicial (Monica debería corregir "I is")
print("Probando respuesta de Monica...")
resp = chatbot_responder("Hello Monica, I is very happy to learn with you.")
print(f"🤖 Monica: {resp}")
```

```python
# ── MILESTONE 3: Historial de conversación ────
historial = []

def chatbot_con_memoria(pregunta_usuario, temperatura=0.7):
    global historial
    historial.append({"role": "user", "content": pregunta_usuario})

    prompt = f"<|begin_of_text|><|start_header_id|>system<|end_header_id|>\n\n{SYSTEM_PROMPT}<|eot_id|>"
    for msg in historial:
        prompt += f"<|start_header_id|>{msg['role']}<|end_header_id|>\n\n{msg['content']}<|eot_id|>"
    prompt += "<|start_header_id|>assistant<|end_header_id|>\n\n"

    respuesta = llamar_modelo(prompt, temperatura)
    historial.append({"role": "assistant", "content": respuesta})
    return respuesta

def reiniciar_chat():
    global historial
    historial = []
    print("✓ Historial de Monica reiniciado.")
```

```python
# ── MILESTONE 4: Loop interactivo ────
reiniciar_chat()
print("═══════════════════════════════════════")
print("  🇺🇸 MONICA - ENGLISH COACH ACTIVADA")
print("  Escribe 'salir' para terminar")
print("═══════════════════════════════════════\n")

while True:
    try:
        entrada = input("👤 Tú: ").strip()
        if entrada.lower() in ["salir", "exit"]:
            print("🤖 Monica: Bye bye! Keep practicing!")
            break
        if not entrada: continue

        respuesta = chatbot_con_memoria(entrada)
        print(f"🤖 Monica: {respuesta}\n")
    except KeyboardInterrupt:
        break
```

```python
# ── MILESTONE 5: Mejoras de Calidad (Estadísticas) ────
def stats_conversacion():
    turnos = len([m for m in historial if m["role"]=="user"])
    print("\n--- 📊 PROGRESO DE LA SESIÓN ---")
    print(f"Intercambios realizados: {turnos}")
    print(f"Mensajes totales en memoria: {len(historial)}")

stats_conversacion()
```
